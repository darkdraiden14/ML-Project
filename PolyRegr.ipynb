```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data loading 
df= pd.read_csv('http://13.234.66.67/summer19/datasets/salary.csv')
```

```python
# experience
exp=df.iloc[:,0:1].values
# sal
sal=df.iloc[:,1:].values
```

```python
# call linear model just for compare
from sklearn.linear_model import LinearRegression
regr1=LinearRegression()
```

```python
# to transform data we can call poly transform method
from sklearn.preprocessing import PolynomialFeatures
```

```python
# experience transform 
poly=PolynomialFeatures(degree=12)
```

```python
# experience fit
newexp=poly.fit_transform(exp)
newsal=poly.fit_transform(sal)
```

```python
# apply linear regression
trained1 = regr1.fit(exp,sal)
```

```python
# now creating linear model again
plt.xlabel('exp')
plt.ylabel('sal')
plt.scatter(exp,sal,label="Actual Salary")
plt.plot(exp,trained1.predict(exp),color="RED",label="linear model predcted salary")
plt.legend()
plt.show()
```

```python
# apply new exp data in model
regr2=LinearRegression()
```

```python
# new exp data apply
trained2=regr2.fit(newexp,sal)
```

```python
# now creating linear model again
plt.xlabel('exp')
plt.ylabel('sal')
plt.scatter(exp,sal,label="Actual Salary")
plt.plot(exp,trained2.predict(newexp),label="new model")
plt.legend()
plt.show()
```
