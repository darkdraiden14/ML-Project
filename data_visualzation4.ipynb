```python
# importing all the required modules
import pandas as pd
import matplotlib.pyplot as plt
import requests
import bs4
```

```python
#fetching data from the given url
url = requests.get('https://en.wikipedia.org/wiki/Machine_learning')
```

```python
soup = bs4.BeautifulSoup(url.text,features='html.parser')
```

```python
# now fetching words
#words = []
#text = soup.get_text()
#for word in text.split():
 #   words.append(word)
words=[]
for text in soup.find_all('p'):
  content=text.text
  words=words+content.split()
  words=list(words)
```

```python
# counting occurence of each word
# for that purpose we will use dict() as we dont know the indexes
wordCount = {}
for word in words:
    if word in wordCount:
        wordCount[word] += 1
    else:
        wordCount[word] = 1

#Sorting the words as per their count
sortWords= []
for key,value in wordCount.items():
    sortWords.append((value,key))

sortWords.sort()
sortWords.reverse()
```

```python
#fetching top 20 words using dataFraming
df=pd.DataFrame(sortWords)
X_count = df.iloc[0:20,0]
Y_words = df.iloc[0:20,1]
```

```python
# plotting pie chart
plt.pie(X_count,labels=Y_words)
plt.show()
```

```python
# plotting bar graph
plt.bar(X_count,Y_words)
```

```python
#Sorting the words as per their count if their count is more than 3
sortWords1= []
for key,value in wordCount.items():
    if value > 3:
        sortWords1.append((value,key))
        
#convert it into dataframe
df1=pd.DataFrame(sortWords1)

X_count1=df.iloc[:100,0]
Y_words1=df.iloc[:100,1]
```

```python
# scatter graph of all the words having presence more than 3 times 
# but i am using top 100 words only as it is not possible to show all words
from matplotlib.pyplot import figure
figure(num=None, figsize=(18, 16), dpi=80, facecolor='w', edgecolor='k')  # just to increase the visibility
plt.scatter(X_count1,Y_words1)
```

```python
# fetching all the links in the page
link =[]
for l in soup.find_all('a',href=True):
    link.append(l['href'])
```

```python
# counting occurence of each link
# for that purpose we will use dict() as we dont know the indexes
linkCount = {}
for l in link:
    if l in linkCount:
        linkCount[l] += 1
    else:
        linkCount[l] = 1

#Sorting the links as per their count
sortlinks= []
for key,value in linkCount.items():
    sortlinks.append((value,key))

sortlinks.sort()
sortlinks.reverse()
```

```python
#fetching top 20 links using dataFraming
df=pd.DataFrame(sortlinks)
X_linkcount = df.iloc[0:10,0]
Y_links = df.iloc[0:10,1]
X_linkcount
```

```python
# stack plot of all the links
plt.stackplot(X_linkcount,Y_links)
```
