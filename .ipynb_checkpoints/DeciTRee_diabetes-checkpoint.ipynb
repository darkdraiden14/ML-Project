```python
# extract attribute from dataframe
features = df.iloc[:,0:8].values
```

```python
features
```

```python
# now label
label=df.iloc[:,8].values
```

```python
# sep training amd testing data
from sklearn.model_selection import train_test_split
```

```python
X,x,Y,y=train_test_split(features,label,test_size=0.2)
# X is training data 
# x is testing data
# Y is training_label / answers
# y is testing_label / answers
```

```python
# callig Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#calling
clf=DecisionTreeClassifier()
```

```python
# now training decisiontree clf
trained=clf.fit(X,Y)
```

```python
#now predicting
predicted=trained.predict(x)
```

```python
predicted
```

```python
y
```

```python
# Check accuracy score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predicted,y)
```

```python
# doing same with KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
kclf=KNeighborsClassifier(n_neighbors=5)
```

```python
ktrain=kclf.fit(X,Y)
```

```python
kpredict=ktrain.predict(x)
```

```python
y
```

```python
accuracy_score(y,kpredict)
```
